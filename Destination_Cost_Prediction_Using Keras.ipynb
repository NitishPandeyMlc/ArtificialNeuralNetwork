## Data

This data is comprised of Destination stay, with 2 measurement features and Destination Cost. Our final goal would be to try to predict the Destination cost of a new destination, in order to try to set a standard cost in the tourism industry.

This having the basic syntax of Keras as Keras is the official API for TensorFlow 2.0.

This is a Regression Problem ,we need to predict the cost of the destination based on the two feature Transportation and hotel_stay (prepared data from internet)

```python
import pandas as pd
import os
```

```python
tdf=pd.read_csv('Destination_cost.csv')
```

```python
tdf.head()
```

```python
tdf.isnull().sum()
```

```python
tdf.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Convert Pandas DF to Numpy array for Keras

X = tdf[['Transportation','Hotel_stay']].values
y = tdf['Destination_Cost'].values

# Train_Test Split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Scailing the Data

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

## Now TensorFlow 2.0 Syntax


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
```

## Model Creation

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation,Dense
```

```python
model=Sequential()

model.add(Dense(3,activation='relu'))
model.add(Dense(3,activation='relu'))
model.add(Dense(3,activation='relu'))
          
## Final Node for Prediction

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')          
```

```python
model.fit(X_train,y_train,epochs=300)
```

```python
#Evaluation

loss = model.history.history['loss']
```

```python
loss
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lineplot(x=range(len(loss)),y=loss)
plt.title('Training Loss Per Epoch')
```

## Evaluation of the Model

```python
#Train Score
train_score=model.evaluate(X_train,y_train)
```

```python
#Test Score
test_score=model.evaluate(X_test,y_test)
```

```python
#Prediction
test_predictions = model.predict(X_test)
```

```python
#Predicted Values
test_predictions
```

```python
#Creating a Dataframe for the Actual Value(Y_test)

Actual_y = pd.DataFrame(y_test,columns=['Actual_y'])
```

```python
Actual_y.head()
```

```python
#Creating a Series as test_predictions was originally in Array

test_predictions = pd.Series(test_predictions.reshape(300,))
```

```python
#Creating a Dataframe where actual and predicted columns exist

pred_dfinal=pd.concat([Actual_y,test_predictions],axis=1)
```

```python
#Renaming the Column Name

pred_dfinal.columns = ['Test_y','Model Predictions']
```

```python
pred_dfinal
```

```python
#Scatter Plot between Actual test and Predicted

sns.scatterplot(x='Test_y',y='Model Predictions',data=pred_dfinal)
```

```python
#To Get the Error (Actual -Predicted)

pred_dfinal['Error'] = pred_dfinal['Test_y'] - pred_dfinal['Model Predictions']
```

```python
#Plotting the Distribution of the Error

sns.distplot(pred_dfinal['Error'],bins=50)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
#Mean Absolute Error
mean_absolute_error(pred_dfinal['Test_y'],pred_dfinal['Model Predictions'])
```

```python
#Mean Squared Error
mean_squared_error(pred_dfinal['Test_y'],pred_dfinal['Model Predictions'])
```

```python
#Test Score which we have calculated Above is the Mean Squared Error

test_score
```
